# Hypothesis Testing

```python
#Test Null Average MonthlyCharges Churn Yes/No Equal
churn.MonthlyCharges.groupby(churn.Churn).mean()
```

```python
from scipy.stats import ttest_ind
```

```python
# Split
cno=churn[churn.Churn=='No']
cyes=churn[churn.Churn=='Yes']
```

```python
ttest_ind(cyes.MonthlyCharges,cno.MonthlyCharges,equal_var=False)
# Since pvalue=8.59244933154705e-73 is less than 0.05,Reject Null
```

```python
#Test Null Average tenure of Churn Yes/No Equal
churn.tenure.groupby(churn.Churn).mean()
```

```python
ttest_ind(cyes.tenure,cno.tenure,equal_var=False)
# Since pvalue=1.1954945472607151e-232 is less than 0.05,Reject Null
```

```python
#Test Null Average Monthly Charges of different PaymentMethod Equal
churn.MonthlyCharges.groupby(churn.PaymentMethod).mean()
```

```python
paybt=churn[churn.PaymentMethod=='Bank transfer (automatic)']
paycd=churn[churn.PaymentMethod=='Credit card (automatic)']
payec=churn[churn.PaymentMethod=='Electronic check']
paymc=churn[churn.PaymentMethod=='Mailed check']
```

```python
from scipy.stats import f_oneway
```

```python
f_oneway(paybt.MonthlyCharges,paycd.MonthlyCharges,payec.MonthlyCharges,paymc.MonthlyCharges)
# Since pvalue=1.1802197193575694e-267 is less than 0.05,Reject Null
```

```python
#Test Null Average tenure of different PaymentMethod Equal
f_oneway(paybt.tenure,paycd.tenure,payec.tenure,paymc.tenure)
# Since pvalue=1.503848361277172e-265 is less than 0.05,Reject Null
```

```python
#Test Null No Asssociation between gender & Churn
from scipy.stats import chi2_contingency
```

```python
chi2_contingency(pd.crosstab(churn.gender,churn.Churn))
# Since pvalue=0.48657873605618596 is less than 0.05,Reject Null
```

```python
#Test Null No Association between SeniorCitizen & Churn 
pd.crosstab(churn.SeniorCitizen,churn.Churn)
```

```python
chi2_contingency(pd.crosstab(churn.SeniorCitizen,churn.Churn))
# Since pvalue=1.510066805092378e-36 is less than 0.05,Reject Null
```
